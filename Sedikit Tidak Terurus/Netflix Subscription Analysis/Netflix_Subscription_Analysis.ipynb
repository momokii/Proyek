```python
data
```

```python
data["mean_price"] = ((data.Basic_month_cost + data.Standar_month_cost + data.Premium_month_cost) / 3).round(2)
data.head()
```

```python
data["cat_price"] = pd.qcut(data.mean_price, [0, 0.25, 0.75, 0.9, 1], labels= ["cheap", "normal", "high", "very_high"])
data.head()
```

```python
data.sort_values("mean_price", ascending  = False).head(10)
```

```python
data.sort_values("mean_price").head(10)
```

```python
data["Library_rank"] = data.Library_size.rank(ascending = False)
data["tvShows_rank"] = data.Tv_shows.rank(ascending = False)
data["Movies_rank"] = data.Movies.rank(ascending = False)
data["mean_price_rank"] = data.mean_price.rank(ascending = False, method = "min")
```

```python
data["Library_cat"] = pd.qcut(data.Library_size, [0, 0.25, 0.75, 0.9, 1], ["less", "avg", "abv_avg", "many"])
```

```python
data.head()
```

```python
data["different_basic_and_premium"] = data.Premium_month_cost - data.Basic_month_cost
```
